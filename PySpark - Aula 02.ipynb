```python
from pyspark.sql.functions import col, asc, desc
```

```python
data = [("John",28),("Anna",24),("Mike",35)]
columns = ["Nome","Idade"]
df = spark.createDataFrame(data, schema= columns)

display(df)
```

### Operações Básicas com DataFrames

##### Selecionando uma coluna

```python
## Selecionando uma coluna
df.select("Nome").display()
```

```python
##filter_df1 = df.filter(df.Idade > 25).display()
##filter_df2 = df.filter(df["Idade"] > 25).display()
filter_df3 = df.filter(col("Idade") > 25).display()
```
