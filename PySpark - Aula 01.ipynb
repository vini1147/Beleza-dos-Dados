#### Começando com PySpark
###### Criando um simples DataFrame

```python
data = [("John",20),("Carlos",39),("Mike",14)]
columns = ["Nome","Idade"]
df = spark.createDataFrame(data,columns)
display(df)
```

```python
## Criando uma tabela a partir de um DataFrame
df.createOrReplaceTempView("tabela")
```

```python
%sql
select * from tabela;
```
