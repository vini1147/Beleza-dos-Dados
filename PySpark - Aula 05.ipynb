### Funções no PySpark
Para realizar operações mais complexas, o PySpark fornece um rico conjunto de funções embutidas. Essas funções permitem que voçê realize operações como agregações, transformações e cálculos em seus dados

#### Funções Comuns
Aqui estão algumas das funções mais comumento usadas no PySpark
- **`col():`** Seleciona uma coluna.
- **`lit():`** Cria um valor literal.
- **`when():`** Cria lógica condicional.
- **`orderBy():`** Agrupa dados por.

```python
from pyspark.sql.functions import col, when, lit
```

```python
# Criando um novo DataFrame
dados = [
    ("John", "Vendas", 3000),
    ("Vinicius", "Analista de Dados", 5000),
    ("Anna", "Marketing", 4000),
    ("Carlos", "Vendas", 3200),
    ("Amanda", "Marketing", 4500),
]
columns = ["Nome","Departamento","Salario"]

df = spark.createDataFrame(dados, schema=columns)
```

```python
display(df)
```

##### Criando uma nova coluna com uma condição usando a função `when` e a `otherwise`

```python
df = (
    df.withColumn("Categoria Salarial",
        when(col("Salario")>= 4500,"Alto")
        .otherwise("Baixo")
    )
)
display(df)
```

##### Agrupando e agregando dados

```python
from pyspark.sql.functions import avg, sum
```

```python
df.groupBy(col('Departamento')).agg(sum('Salario').alias("Soma Salarial")).display()
```

##### Remover linhas em branco com função `dropna`

```python
# Criando um novo DataFrame
dados = [
    ("John", "Vendas", 3000),
    ("Vinicius", "Analista de Dados", 5000),
    ("Anna", "Marketing", 4000),
    ("Carlos", "Vendas", 3200),
    ("Amanda", "Marketing", 4500),
    ("João", None, 2000),
    ("Caio", None, 4200),
    ("Julia", "", 1000),
    ("Caio", " ", 6200)
]
columns = ["Nome","Departamento","Salario"]

df = spark.createDataFrame(dados, schema=columns)

display(df)
```

```python
df = df.dropna()
display(df)
```

```python
# Remove a linha inteira apenas se todas as colunas estiverem vazias
df = df.dropna(how='all').display()
```

##### Removendo Coluna com a função `drop()`

```python
##df = df.withColumn("nula",lit(None))
df = df.drop("")
df.display()
```

##### Preenchendo valores nulos com a função `fillna()`

```python
df = df.fillna("Desconhecido", subset=["Departamento"])
df.display()
```
