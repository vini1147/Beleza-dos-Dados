```python
from pyspark.sql.functions import col, desc, asc
```

```python
data = [("John",28),("Anna",24),("Mike",35)]
columns = ["Nome","Idade"]
df = spark.createDataFrame(data, schema= columns)
```

```python
# Adciona uma coluna nova a partir da coluna "Idade" e soma 5 a ela
df = df.withColumn("Idade dps de 5 anos",col("Idade") + 5)
display(df)
```

```python
transformed_df = df.select("Nome", "Idade").filter(df["Idade"]> 25)
transformed_df.collect()
```

```python
# Criando um novo DataFrame
dados = [
    ("John", "Vendas", 3000),
    ("Vinicius", "Analista de Dados", 5000),
    ("Anna", "Marketing", 4000),
    ("Carlos", "Vendas", 3200),
    ("Amanda", "Marketing", 4500),
]
columns = ["Nome","Area","Salario"]
df_funcionarios = spark.createDataFrame(dados, schema = columns)
display(df_funcionarios)
```

```python
# Filtrando funcionários do departamento de Vendas
df_funcionarios_vendas = df_funcionarios.filter(df_funcionarios["Area"] == "Vendas")
display(df_funcionarios_vendas)
```

```python
# Calculando a média do Salário agrupando com Área
media_salario = df_funcionarios.groupBy("Area").avg()
display(media_salario)
```
